```python
# Algorithmic Assignment: Plot Chessboard
chessboard = np.zeros((8,8), dtype='<U1')

# chessboard[1::2, ::2] = 1
# chessboard[::2, 1::2] = 1
black = '\u2B1B'
white = '\u2B1C'
for row in range(8):
  for column in range(8):
    if(row%2 == 0 and column%2 == 1)or (row%2 == 1 and column%2 == 0):
      chessboard[row, column] = white
    else:
      chessboard[row, column] = black

# Pyhton list
chessboard=chessboard.tolist()


# print(chessboard)
# print(type(chessboard))
for row in chessboard:

  print(row)
```

```python
scores = np.array([
                    [80, 85, 90],
                    [70, 75, 80],
                    [90, 95, 100],
                    [60, 65, 70]
                  ])
print(scores)
print(scores.shape)
```

```python
another_scores = np.array([
                    [80, 85, 90],
                    [60, 65, 70]
                  ])

print(another_scores)
print(another_scores.shape)
```

```python
more_scores = np.array([
                    [80],
                    [60],
                    [115],
                    [97]
                  ])

print(more_scores)
print(more_scores.shape)
```

```python
# Stacking Arrays -> Concatenating Arrays
result1 = np.vstack((scores, another_scores))
# vertical stack -> no of columns should match
print(result1)
print(result1.shape)
```

```python
result2 = np.hstack((scores, more_scores))
# horizontal stacking -> no of rows should match
print(result2)
print(result2.shape)
```

```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image_path = 'husky.jpeg'
image = Image.open(image_path)
image_array = np.array(image)
print(image_array)
print(image_array.shape)
```

```python
# plt.imshow(image)
plt.imshow(image_array)
# red green blue

```

```python
print(image_array[0][0])
```

```python
# Assignment: Greyscale i.e. Black And White
pixel_value = (25+34+31)//3
image_array[0][0][0] = pixel_value
image_array[0][0][1] = pixel_value
image_array[0][0][2] = pixel_value

plt.imshow(image_array)
```

```python
# Assignment: Greyscale i.e. Black And White
pixel_value = (132+150+170)//3
image_array[0][0][0] = pixel_value
image_array[0][0][1] = pixel_value
image_array[0][0][2] = pixel_value

rotated_image_array = np.rot90(image_array, k=1)

plt.imshow(rotated_image_array)
```

```python
columns_to_remove = [1, 100]
cropped_image_array = np.delete(image_array, columns_to_remove, axis=1)
print(cropped_image_array)
print(image_array)
plt.imshow(image_array)
```

```python
# Assignment: Greyscale i.e. Black And White
# for loop
height, width, _ = image_array.shape
for row in range(height):
    for col in range(width):
        # Calculate the average grayscale value
        pixel_value = np.sum(image_array[row, col]) // 3

        # Assign the same grayscale value to all three channels (R, G, B)
        image_array[row, col] = [pixel_value, pixel_value, pixel_value]

print(image_array)
plt.imshow(image_array)
```
